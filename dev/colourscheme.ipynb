```python
# import oscovida
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']

import numpy as np
```

```python
def create_demo_figure(colours=None, save=False):
    fig, ax = plt.subplots(figsize=(10, 6))
    N = 20
    x = np.linspace(0, 1, N)
    
    if colours:
        pass
    else:
        colours = range(10)
    
    for i, label in enumerate(colours):
        color = "C" + str(label)
        y = np.sin(x*2*np.pi)*i/N + 1.2 * i/N
        ax.plot(x, y, 'o-', color=color, label=f'{color}', linewidth=10)
        ax.legend()

    if save:
        fig.savefig('colourscheme.pdf')

    
```

```python
plt.style.use('default')
create_demo_figure()
create_demo_figure([0, 1, 2, 3, 7, 8])
```

```python
plt.style.use('ggplot')
create_demo_figure()
create_demo_figure([0, 1, 2, 3, 4, 5])
```

```python
plt.style.use('fivethirtyeight')
create_demo_figure()
create_demo_figure([0, 1, 2, 3, 4, 5])
```
